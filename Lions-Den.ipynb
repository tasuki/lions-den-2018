## 2. Photos taken abroad vs account balance

```python
df[df.SM_tags == -9999]
```

```python
# get those with deposit accounts and without invalid number of tagged pictures
tags_balance = df[["SM_tags", "balance"]][(df.deposit == 1) & (df.SM_tags >= 0)].sort_values(by="SM_tags")
# get avg account balance per number of tags
grouped = tags_balance.groupby("SM_tags").agg(["mean", "count"])
grouped.columns = ["mean_balance", "client_count"]
grouped.head(15)
```

```python
grouped.plot(logx=True)
pass
```

#### Conclusions:

1. The data is noisy: there are five people with -9999 outside tagged photos: that's clearly wrong. Curiously, all of them have balance over 10k. Verify the data collection procedures.
1. There very much is a dependency between number of tagged photos and the account balance.
1. The number of tagged photos correlates positively with account balance up till about 10 photos. The group with 6-10 tagged pictures is the only one that has more than 5k balance on average. Theory: the more money one has, the more they can afford foreign travel.
1. People with more than 13 tagged photos don't have very high account balance. Also, there isn't a whole lot of them. Theory: These are the people who spend all their money on travelling.

## 3. Life insurance

```python
df.corr()["action"]
```

Looks like the number of social media accounts and number of friends are most strongly correlated with whether or not the person has taken a life insurance, balance is much less correlated, and the other fields even less so.

```python
accounts_insurance = df[["SM_account_number", "action"]].fillna(-1).sort_values("SM_account_number")
accounts_insurance.groupby("SM_account_number").agg(["mean", "count"])
```

Yep, number of social media accounts is highly negatively correlated with life insurance.

```python
friends_insurance = df[["SM_friends", "action"]].fillna(-1).sort_values("SM_friends")
friends_insurance = friends_insurance.groupby("SM_friends").agg(["mean", "count"])
friends_insurance.columns = ["insured", "count"]

base = 50.0
friends_insurance["smooth"] = friends_insurance.index
friends_insurance["smooth"] = friends_insurance["smooth"].apply(lambda x: base * round(x / base))

friends_insurance = friends_insurance.groupby("smooth").agg(["mean", "sum"])

friends_insurance.columns = ["insured", "insured_sum", "count_mean", "users"]
friends_insurance = friends_insurance.drop(columns=["insured_sum", "count_mean"])
```

```python
cmap = matplotlib.cm.get_cmap('tab20c')
color1 = cmap(0.0)
color2 = cmap(0.2)

fig, ax = plt.subplots()
ax1, ax2 = ax, ax.twinx()

ax1.set_xlabel("friends")
ax1.set_ylabel("insured", color=color1)
ax1.tick_params(axis='y', colors=color1)
ax1.plot(friends_insurance["insured"], label="insured", color=color1)

ax2.set_ylabel("clients within range", color=color2)
ax2.tick_params(axis='y', colors=color2)
ax2.plot(friends_insurance["users"], label="clients", color=color2)

plt.show()
```

And also number of social media friends is highly negatively correlated.

```python
df.groupby("deposit").agg(["mean", "count"])["action"]
```

Life insurance bought by 11% of non-account holders and 15% for account holders, not much of a difference. Let's further look at the account holders...

```python
balance_insurance = df[["balance", "action"]][df.deposit == 1].sort_values("balance")
balance_insurance.groupby("balance").agg(["mean", "count"])

base = 500.0
balance_insurance["smooth"] = balance_insurance["balance"].apply(lambda x: base * round(x / base))
balance_insurance = balance_insurance.drop(columns=["balance"])
balance_insurance = balance_insurance.groupby("smooth").agg(["mean", "count"])
balance_insurance.columns = ["insured", "count"]
print(balance_insurance)

ax1 = balance_insurance["insured"].plot(color=color1)
ax1.set_xlabel("account balance")
ax1.tick_params(axis='y', colors=color1)
ax1.set_ylabel("insured", color=color1)

ax2 = ax1.twinx()
ax2.tick_params(axis='y', colors=color2)
ax2.set_ylabel("clients within range", color=color2)
ax2.plot(balance_insurance["count"], color=color2)
pass
```

Account balance is weakly correlated with buying insurance.

#### Conclusions:

1. An average person is ~15% likely to buy life insurance. People who only have 1 or 2 social media accounts are almost 30% likely to buy life insurance, as opposed to people with 3 or 4 social media accounts, of whom less than 1% buy life insurance.
1. Furthermore, people in the "first hump" with less than 300 friends are more likely to buy life insurance, while people in the "second hump" with more than 300 friends are not very likely at all to buy life insurance. The number of sm accounts is highly correlated with the number of sm friends, so this result is perhaps unsurprising.
1. The obvious explanation for this occurence is that life insurance is much less prevalent among young people, who are heavy social media users.
1. Account balance is positively correlated with buying life insurance. The correlation is rather weak. Furthermore, there's a lot of noise in the account balances above 7k where there are fewer people. Perhaps richer people can afford life insurance, or perhaps older people who buy more life insurance are richer.
